# catagorical plot (catplot())

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.catplot(x='sex',y='total_bill',data=df)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,jitter=0.1)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='swarm',hue='smoker')
plt.show()
```

# Box  and Boxen plots

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='box')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='box',hue='smoker')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='boxen',hue='smoker')
plt.show()
```
