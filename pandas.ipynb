```python
import pandas as pd
columns = ['sl','sw','pl','pw','Species']
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=columns)

df = iris.copy()
df = df[df.Species == "Iris-setosa"]
df = df.describe()
# print(df)

min = df.iloc[3]
# print(min)
for i in min :
    print(format(i,'.2f'),end = " ")
print("Iris-setosa")
max = df.iloc[7]
for i in max :
    print(format(i,'.2f'),end = " ")
print("Iris-setosa")
avg = df.iloc[1]
for i in avg :
    print(format(i,'.2f'),end = " ")
print("Iris-setosa")

df = iris.copy()
df = df[df.Species == "Iris-versicolor"]
df = df.describe()
# print(df)
min = df.iloc[3]
for i in min :
    
    print(format(i,'.2f'),end = " ")
print("Iris-versicolor")
max = df.iloc[7]
for i in max :
    print(format(i,'.2f'),end = " ")
print("Iris-versicolor")
avg = df.iloc[1]
for i in avg :
    print(format(i,'.2f'),end = " ")
print("Iris-versicolor")

df = iris.copy()
df = df[df.Species == "Iris-virginica"]
df = df.describe()
# print(df)
min = df.iloc[3]
for i in min :
    print(format(i,'.2f'),end = " ")
print("Iris-virginica")
max = df.iloc[7]
for i in max :
    print(format(i,'.2f'),end = " ")
print("Iris-virginica")
avg = df.iloc[1]
for i in avg :
    print(format(i,'.2f'),end = " ")
print("Iris-virginica")
```
